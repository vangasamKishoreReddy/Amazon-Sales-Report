##  1. Univariate analysis

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Univariate analysis - Numeric columns (Qty, Amount)
sns.histplot(df['Qty'], kde=True)
plt.title('Distribution of Quantity')
plt.show()

sns.histplot(df['Amount'], kde=True)
plt.title('Distribution of Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=df['ship-service-level'], order=df['ship-service-level'].value_counts().index)
plt.title("Shipping Service Level Distribution")
plt.xlabel("Count")
plt.ylabel("Ship Service Level")
plt.show()
```

```python
# Univariate analysis - Categorical columns (Sales Channel, Category)
sns.countplot(x='Sales Channel', data=df)
plt.title('Sales Channel Distribution')
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='Category', data=df)
plt.title('Category Distribution')
plt.xticks(rotation=45)
plt.show()

sns.countplot(x='ship-country', data=df)
plt.title('Ship-Country Distribution')
plt.xticks(rotation=45)
plt.show()
```

# 2.  Bivariate Analysis

```python
plt.figure(figsize=(12, 6))
df['Date'] = pd.to_datetime(df['Date'])  # Convert to datetime format
df.groupby(df['Date'].dt.date)['Order ID'].count().plot(kind='line', marker='o')
plt.title("Orders Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Bivariate analysis - Scatter Plot
sns.scatterplot(x='Qty', y='Amount', data=df)
plt.title('Relationship Between Quantity and Amount')
plt.show()
```

```python
# Bivariate analysis - Heatmap
correlation_matrix = df[['Qty', 'Amount']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Qty and Amount')
plt.show()
```

```python
# Bivariate analysis - Box Plot (Amount vs Sales Channel)
sns.boxplot(x='Sales Channel', y='Amount', data=df)
plt.title('Amount vs Sales Channel')
plt.xticks(rotation=45)
plt.show()
```

# 3. Multivariate Analysis:

```python
# Multivariate analysis - Pair Plot
sns.pairplot(df[['Qty', 'Amount']])
plt.show()
```

```python
# Convert columns to numeric, replacing errors with NaN
df['Qty'] = pd.to_numeric(df['Qty'], errors='coerce')
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['ship-postal-code'] = pd.to_numeric(df['ship-postal-code'], errors='coerce')

# Now calculate the correlation matrix
correlation_matrix = df[['Qty', 'Amount', 'ship-postal-code']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
